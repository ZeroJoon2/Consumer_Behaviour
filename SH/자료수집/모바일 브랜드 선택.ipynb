# [ 모바일 브랜드 선택 요인 ]

## < "모바일 데이터서비스 선택 결정요인에 관한 연구" >

### 1. 요약
(1) 스마트폰과 LTE 서비스 확산으로 소비자들이 모바일 데이터 서비스를 선택했을때 어떤 요인들을 고려한지 분석하였음.   

(2)전송속도, 데이터 제공량, 무제한 요금재 등의 서비스 품질 요소에 대한 어떤 영향을 미치는지에 대해 다룸 

### 2. 브랜드 선택 결정 요인
(1) 주요 요인
 - 전송속도: LTE 서비스 도입 후 빠른 전송 속도가 중요해짐  
 - 데이터 제공량: 제공량에 따라 소비자 선택이 달라짐(요금제에도 영향을 줌)  
 - 가격(요금제): 가격 수준이 가장 중요함, 요금 낮을수록 선호도 증가  
 - 부가서비스: 데이터 쉐어링, mVoIP 지원 여부, 음악·영상 전용 요금제
 
(2) 서비스별 중요도  
 - 1순위: 요금수준(47.1%)    
 - 2순위: 데이터 기본 제공량(29.8)  
 - 3순위: 전송속도(23%)  
 
(3) 사용자별 중요도  
 - LTE 사용자: 데이터 기본제공량, 전송속도  
 - 3G 사용자: 요금 수준이 절대적 중요  

## < "모바일 서비스 사용의 영향 요인에 관한 연구" >

### 1. 요약
(1) 모바일 서비스 사용을 미치는 요인들을 문헌 연구에서 추출하여서 알려주고 있음.

(2)소비자들이 모바이 서비스 선택하고 사용하는데 영향을 주는 요인에 대해 파악가능

### 2. 브랜드 선택 요인
(1) 주요 요인
 - 제품: 디자인, 성능, 기능
 - 서비스: A/S, 보증 서비스, 고객 대응
 - 모바일서비스: 네트워크 품질, 데이터 제공량, 앱 및 생태계 

(2) 국가별 차이점  
- 한국: 제품, 서비스 선호도가 높음  
- 중국: 제품, 모바일 속성 선호도가 높음  
-> 문화에 따라 고려요인이 달라질 수 있음
